```python
import pandas as pd
from modulku import praproses as pps, StemNstopW as stm
from replace import normalisasi as nrm
```

```python
data = pd.read_excel("komentar/gojek_twitter_dataset.xlsx")
komentar = data.tweet.tolist()
label = data.sentimen.tolist()
data.head()
```

```python
data_bersih = list()
label_data_bersih = list()
lama = list()
for ix, i in enumerate(komentar):
    if type(i) is float:
        i = str(i)
    i = pps.preprocessing(i)
    i = nrm.normalisasi(i)
    i = stm.stemmer_kata(i)
    i = stm.stop_word(i)
    
    if len(i)>3:
        data_bersih.append(i)
        label_data_bersih.append(label[ix])
        lama.append(komentar[ix])
    if (ix+1)%100==0:
        print(ix+1)
    else:
        print('.', end='')
```

```python
dictt = {
#     "nomer":data.Column1.tolist(),
    'label':label_data_bersih,
    'komentar':data_bersih, 
    'komentar_lama':lama
}
df = pd.DataFrame.from_dict(dictt)
df.to_excel('komentar/data_clean_goejk.xlsx')
```
